```python

import numpy as np

#Q1

#a
v=np.array([2,3.8,0,5,1])
print("Q1. a")
print(v,"\n")

#b
u=np.arange(1,11,2).reshape(5,1)
u=np.exp(u)
print("Q1. b")
print(u,"\n")

#c
w=np.arange(-10,4)
print("Q1. c")
print(w,"\n")

#d
m=np.arange(-3,9,2)
print("Q1. d")
print(m,"\n")


#e
n=np.arange(0,8,2)
print("Q1. e")
print(n,"\n")


#f
s=np.arange(0,-3.1,-0.1).reshape(31,1)
print("Q1. f")
print(s,"\n")


#g
t=np.linspace(0,np.pi,50).reshape(50,1)
print("Q1. g")
print(t,"\n")

```

```python
#Q2

#a
A=np.array([[1,2,3],[5,3,1],[7,7,7]])
print("Q2. a")
print(A,"\n")

#b
tmp=np.zeros((5,5))
print("Q2. b")
print(tmp,"\n")

#c
B=2*np.ones((5,5))
print("Q2. c")
print(B,"\n")

#d
C=5*np.ones((5,5))-5*np.identity(5)
print("Q2. d")
print(C,"\n")

#e
D=np.diag([1,2,3,2,1])
print("Q2. e")
print(D,"\n")

#f
F=np.ones((9,5))*np.arange(1,10).reshape(9,1)
print("Q2. f")
print(F,"\n")

```
